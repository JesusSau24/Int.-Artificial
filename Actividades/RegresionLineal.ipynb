```
import matplotlib.pyplot as plt # generar gráficos
import numpy as np # generar arreglos
from sklearn.linear_model import LinearRegression# contiene diferentes modelos de ML
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```
np.random.seed(0)
horas_estudio = np.random.rand(100,1)*10
calificaciones = 2*horas_estudio + np.random.rand(100, 1)*2
```

```
calificaciones_train, calificaciones_test, horas_estudio_train, horas_estudio_test = train_test_split(calificaciones, horas_estudio, test_size=0.2, random_state=20)
```

```
modelo_regresion = LinearRegression()
modelo_regresion.fit(calificaciones_train, horas_estudio_train)
```

```
predicciones = modelo_regresion.predict(calificaciones_test)
```

```
plt.scatter(calificaciones, horas_estudio, color='red')
plt.xlabel("Calificacion obtenida")
plt.ylabel("Tiempo de estudio en horas")
plt.title("GRAFICA DE REGRESION LINEAL")
plt.ylim(0, 11)
plt.plot(calificaciones, modelo_regresion.predict(calificaciones), color='blue')
```

```
error_cuadratico_medio = mean_squared_error(horas_estudio_test, predicciones)
print("Error cuadrático medio:", error_cuadratico_medio)
```
